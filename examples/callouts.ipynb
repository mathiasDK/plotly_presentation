```python
from plotly_presentation import Plotter
```

```python
import plotly.express as px
df = px.data.stocks()
p = Plotter()
p.express(
    type="line",
    data_frame=df,
    x="date",
    y=["GOOG", "AAPL", "FB"]
)
p.callout.add_line_end_marker(
    text_type="category"
)
p.show()
```

```python
p = Plotter(slide_layout="slide_50%")
p.express(
    type="line",
    x=["cat1", "cat1", "cat2", "cat2", "cat3", "cat3", "cat4", "cat4"],
    y=[1, 3, 2, 3, 3, 2, 4, 3],
    color=["a", "b", "a", "b", "a", "b", "a", "b"],
)
p.callout.add_line_end_marker(
    text_type="category",
    text_positions=["middle right", "top right"]
)
p.show()
```

```python
p = Plotter(slide_layout="slide_50%")
p.express(
    type="bar",
    x=["cat1", "cat1", "cat2", "cat2", "cat3", "cat3", "cat4", "cat4"],
    y=[1, 3, 2, 3, 3, 2, 4, 4],
    color=["a", "b", "a", "b", "a", "b", "a", "b"],
    barmode="group"
)
p.callout.add_line_differences(
    primary_trace_name="b",
    secondary_trace_name="a",
    text_type="ratio",
    text_format=".1f",
    y_offset=0.1
)
p.show()
```

```python
p = Plotter(slide_layout="slide_50%")
p.express(
    type="bar",
    x=[1,2,3],
    y=[1,2,3],
)
p.callout.add_circle_highlight(x=2, y=3, text="1")
p.show()
```

```python
p = Plotter(slide_layout="slide_50%")
p.express(
    type="bar",
    x=[1,2,3],
    y=[1,2,3],
)
p.callout.add_square_growth_line(
    x0=1, x1=3, y0=1.1, y1=3.1, y_top=3.5, 
    # circle_x_pixel_width=35, 
    # circle_y_pixel_width=18, 
    text="+20%"
    )
p.show()
```

```python
p = Plotter(slide_layout="slide_50%")
p.express(
    type="bar",
    x=[1,2,3],
    y=[1,2,3],
)
p.callout.add_dash_growth_lines(x0=1, x1=3, y0=1, y1=3, x_end=3.7, text="+20%")#, circle_x_pixel_width=35, circle_y_pixel_width=18)
fig = p.figure
fig.show()
```
