```python
from plotly_presentation import Plotter
import plotly_presentation
import plotly.express as px
```

```python
p = Plotter()
p.express(
    type="line",
    data_frame=px.data.stocks(),
    x="date",
    y=["GOOG", "FB", "AAPL"],
    # color_continuous_scale=c.sequential.Sequential
)
p.show()
```

```python
p.style.set_color_palette(palette_type="diverging", palette_name="reds")
p.show()
```
